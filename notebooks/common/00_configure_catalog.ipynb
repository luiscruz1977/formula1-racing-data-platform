```python
# COMMAND ----------
# Project Configuration

CATALOG = "formula1_racing"

BRONZE_SCHEMA = "bronze"
SILVER_SCHEMA = "silver"
GOLD_SCHEMA = "gold"

RAW_VOLUME = "raw_files"

RAW_PATH = f"/Volumes/{CATALOG}/{BRONZE_SCHEMA}/{RAW_VOLUME}"
```

```python
# COMMAND ----------
# Helper Functions

def execute_sql(sql_statement: str):
    """
    Execute a SQL statement and print it to the notebook output.
    """
    print(f"Executing:\n{sql_statement}\n")
    spark.sql(sql_statement)
```

```python
# COMMAND ----------
# Create Schemas

for schema in [BRONZE_SCHEMA, SILVER_SCHEMA, GOLD_SCHEMA]:

    execute_sql(f"""
        CREATE SCHEMA IF NOT EXISTS {CATALOG}.{schema}
    """)
```

```python
# COMMAND ----------
# Create Bronze Volume

execute_sql(f"""
CREATE VOLUME IF NOT EXISTS
{CATALOG}.{BRONZE_SCHEMA}.{RAW_VOLUME}
""")
```

```python
# COMMAND ----------
# Validate Schemas

display(
    spark.sql(f"SHOW SCHEMAS IN {CATALOG}")
)
```

```python
# COMMAND ----------
# Validate Volumes

display(
    spark.sql(f"SHOW VOLUMES IN {CATALOG}.{BRONZE_SCHEMA}")
)
```

```python
# COMMAND ----------
# Configuration Summary

print("=" * 50)
print("Formula 1 Racing Data Platform")
print("=" * 50)

print(f"Catalog       : {CATALOG}")
print(f"Bronze Schema : {BRONZE_SCHEMA}")
print(f"Silver Schema : {SILVER_SCHEMA}")
print(f"Gold Schema   : {GOLD_SCHEMA}")
print(f"Raw Path      : {RAW_PATH}")

print("=" * 50)
```
